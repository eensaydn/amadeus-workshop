```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

def generate_email(scenario, additional_info=""):
    system_prompt = """You are a professional business communication expert.
Based on the scenario:
- Use polite and professional tone
- Keep it brief (max 150 words)
- Use appropriate opening and closing
- Follow proper grammar rules"""

    user_prompt = f"""Scenario: {scenario}
Additional Info: {additional_info if additional_info else 'None'}

Write an appropriate email for this scenario."""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0.7,
        max_tokens=400
    )
    return response.choices[0].message.content

# Meeting invitation email
email = generate_email(
    scenario="Team meeting invitation",
    additional_info="Date: November 20, 2025, Time: 2:00 PM, Location: Zoom, Topic: Q4 Review"
)

print("="*70)
print("MEETING INVITATION EMAIL")
print("="*70)
print(email)
```
