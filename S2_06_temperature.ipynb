```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

def generate_with_temperature(prompt, temp):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "user", "content": f"{prompt}\n\nKISA VE OZ CEVAP VER (maksimum 3-4 cumle)."}
        ],
        temperature=temp,
        max_tokens=100  # DÃ¼ÅŸÃ¼rÃ¼ldÃ¼ - kÄ±sa cevaplar iÃ§in
    )
    return response.choices[0].message.content

# Test prompts
test_cases = [
    
    {
        "prompt": "Bir kahve dukkani icin yaratici slogan uret.",
        "type": "YARATICI (farklilik onemli)"
    },

]

for i, case in enumerate(test_cases, 1):
    print(f"\n{'='*70}")
    print(f"TEST {i} ({case['type']}): {case['prompt']}")
    print('='*70)
    
    print("\nTEMPERATURE 0.1 (Dusuk - Tutarli):")
    print(generate_with_temperature(case['prompt'], 0.1))
    
    print("\nTEMPERATURE 0.7 (Orta - Dengeli):")
    print(generate_with_temperature(case['prompt'], 0.7))
    
    print("\nTEMPERATURE 1.5 (Yuksek - Yaratici):")
    print(generate_with_temperature(case['prompt'], 1.5))
    
    print("-"*70)  # AyÄ±rÄ±cÄ± ekledim
```
