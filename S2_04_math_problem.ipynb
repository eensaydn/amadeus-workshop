```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

# NORMAL (WITHOUT CoT)
def solve_normal(problem):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "user", "content": problem}
        ],
        temperature=0.1,
        max_tokens=300
    )
    return response.choices[0].message.content

# CHAIN-OF-THOUGHT
def solve_cot(problem):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "user", "content": f"{problem}\n\nAdim adim dusunerek cozelim:"}
        ],
        temperature=0.1,
        max_tokens=300
    )
    return response.choices[0].message.content

# Test problems
problems = [
                "AyÅŸeâ€™nin yaÅŸÄ±, Mehmetâ€™in yaÅŸÄ±nÄ±n iki katÄ±ndan 5 eksiktir. Mehmet 12 yaÅŸÄ±nda olduÄŸuna gÃ¶re AyÅŸe, Mehmetâ€™in doÄŸduÄŸu gÃ¼n kaÃ§ yaÅŸÄ±ndaydÄ±?"
]

for i, problem in enumerate(problems, 1):
    print(f"\n{'='*70}")
    print(f"PROBLEM {i}: {problem}")
    print('='*70)
    
    print("\nNORMAL CEVAP (direkt):")
    print(solve_normal(problem))
    
    print("\nCHAIN-OF-THOUGHT (adim adim):")
    print(solve_cot(problem))
```
