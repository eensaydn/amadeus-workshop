```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))
```

```python
# WITHOUT SYSTEM PROMPT
def answer_without_role(question):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "user", "content": question}
        ],
        temperature=0.7,
        max_tokens=200
    )
    return response.choices[0].message.content
```

```python
# WITH SYSTEM PROMPT (ROLE)
def answer_with_role(question, role):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": role},
            {"role": "user", "content": question}
        ],
        temperature=0.7,
        max_tokens=200
    )
    return response.choices[0].message.content
```

```python
# Test scenarios - CIKTI STILI COK FARKLI OLACAK ROLLER
scenarios = [
    {
        "question": "Yeni bir proje nasil baslatilir?",
        "role1": "Sen 5 yasindaki bir cocuga ogretmenlik yapan sabir ve sevgi dolu bir ogretmensin. Her seyi basit orneklerle anlat.",
        "role2": "Sen asiri teknik detaylara giren, sadece kod ve komutlarla konusan bir Linux sistem yoneticisisin."
    },
    
]

for i, scenario in enumerate(scenarios, 1):
    print(f"\n{'='*70}")
    print(f"SENARYO {i}: {scenario['question']}")
    print('='*70)
    
    print("\nSYSTEM PROMPT YOK (normal):")
    print(answer_without_role(scenario['question']))
    
    print(f"\n{'='*70}")
    print(f"ROL 1: {scenario['role1']}");
    print("-"*70)
    print(answer_with_role(scenario['question'], scenario['role1']))
    
    print(f"\n{'='*70}")
    print(f"ROL 2: {scenario['role2']}")
    print("-"*70)
    print(answer_with_role(scenario['question'], scenario['role2']))
```
