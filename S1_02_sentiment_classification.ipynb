```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

def analyze_sentiment(review):
    prompt = f"""Analyze customer reviews and classify as "POSITIVE" or "NEGATIVE" only.

EXAMPLES:
Review: "Great product, very satisfied!"
Result: POSITIVE

Review: "Terrible experience, never again."
Result: NEGATIVE

Review: "Good for the price but didn't meet expectations."
Result: NEGATIVE

Review: "Fast shipping and quality packaging, thanks!"
Result: POSITIVE

NOW CLASSIFY THIS REVIEW:
Review: "{review}"
Result:"""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1,
        max_tokens=10
    )
    return response.choices[0].message.content.strip()

# Test reviews
test_reviews = [
    "Product doesn't deliver, waste of money.",
    "Excellent quality, highly recommend!",
    "Arrived late but product is good.",
    "Customer service is terrible, never buying again.",
    "Great value for money!"
]

for i, review in enumerate(test_reviews, 1):
    result = analyze_sentiment(review)
    emoji = "âœ…" if "POSITIVE" in result.upper() else "âŒ"
    print(f"{i}. {emoji} {review}")
    print(f"   Result: {result}\n")
```
